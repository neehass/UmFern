```python
import earthaccess
print(earthaccess.__version__)
print(earthaccess.__file__)

dir = "C:/Users/maiim/Documents/25-26WS/Umweltfernerkundung/Phosphate"

# chat gpt 
ea = earthaccess
```

```python

ea.login(persist=True)

results = ea.search_data(
    short_name="MODIST_L3m_CHL",                  # Terra MODIS Level‑3 Chlorophyll
    granule_name="*09*",                          # nur September (Monat 09)
    temporal=("2000-09-01", "2025-09-30")
)

ea.download(results, dir + "/MODIS_SEPTEMBER_DATA")
```

```python
# Terra MODIS Level-2 Ocean Color (OC)
# MODIST_L2_OC_NRT
# TSM aus Ocean Color Reflektanz (Rrs) berechnen
# ! Ocean color oft ungenau in Küstenregion >> trotzdem probieren 
ea.login(persist=True)

results = ea.search_data(
    short_name="MODIST_L2_OC", # "MODIST_L2_OC_NRT",
    granule_name="*09*",
    temporal=("2000-09-01", "2025-09-30")
)

ea.download(results, dir +"/MODIS_SEPTEMBER_DATA_OceanColor_2")
```

```python
# PIC (Particulate Inorganic Carbon)
# zb Calciumcarbonat (CaCO₃)
# entsteht durch biologische Kalkbildung (algen) 
# Hyptohese: Indsutrieabwässer >> eutrophierun >> mehr algen oder veränderung der alkalinität und amit mehr PIC

ea.login(persist=True)

results = ea.search_data(
    short_name="MODIST_L3m_PIC", # "MODIST_L3m_PIC_NRT
    granule_name="*09*",
    temporal=("2000-09-01", "2025-09-30")
)

ea.download(results, dir + "/MODIS_SEPTEMBER_DATA_PIC_2")
```
